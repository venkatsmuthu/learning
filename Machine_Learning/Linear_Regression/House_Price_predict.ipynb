```python
dataset = pd.read_csv('House Price Prediction Dataset.csv')
```

```python
dataset = dataset.drop(['Id'], axis='columns')
dataset.head(5)
```

```python
print(dataset['Condition'].unique())
print(dataset['Location'].unique())
print(dataset['Garage'].unique())
```

```python
dataset['Garage'] = dataset['Garage'].map({'Yes':1,'No':0}).astype(int)
dataset['Condition'] = dataset['Condition'].map({'Excellent':4,'Fair':3,'Good':2,'Poor':1}).astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['Location'] = le.fit_transform(dataset['Location'])
```

```python
X = dataset.drop(['Price'], axis='columns')
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = pd.DataFrame(sc.fit_transform(X_train))
X_test = pd.DataFrame(sc.transform(X_test))
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
pred_y = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, pred_y)
```
